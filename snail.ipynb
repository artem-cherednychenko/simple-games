# Snail

Below I present my solution of codewars's eponymous kata ([link](https://www.codewars.com/kata/521c2db8ddc89b9b7a0000c1/python)).

Task: to create a function taking an **n x n** array and return the array elements arranged from outermost elements to the middle element, traveling clockwise.

See the image illustrating the task more clearly:

*NOTE 1*: The idea is not sort the elements from the lowest value to the highest; the idea is to traverse the 2-d array in a clockwise snailshell pattern.

*NOTE 2*: The 0x0 (empty matrix) is represented as en empty array inside an array [[]].

```python
import numpy as np

def snail(arr):
    
    # Create empty 1D array for further output
    snail_arr = np.empty(0)
    
    while True:
        try:         
            # Go right to add top row to snail array, then delete it
            snail_arr = np.concatenate((snail_arr,arr[0]))
            arr = np.delete(arr,0,0)
            # Go down to add rightmost column to snail array, then delete it
            snail_arr = np.concatenate((snail_arr,arr[:,-1]))
            arr = np.delete(arr,-1,1)
            # Go left to add reversed bottom row to snail array, then delete it
            snail_arr = np.concatenate((snail_arr,np.flip(arr[-1])))
            arr = np.delete(arr,-1,0)
            # Go up to add reversed leftmost column to snail array, then delete it
            snail_arr = np.concatenate((snail_arr,np.flip(arr[:,0])))
            arr = np.delete(arr,0,1)
        except:
            break
    
    return snail_arr.astype(int).tolist()
```
